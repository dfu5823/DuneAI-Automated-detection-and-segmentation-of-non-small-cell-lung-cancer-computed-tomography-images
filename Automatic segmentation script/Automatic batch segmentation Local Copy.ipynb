```python
###################################################################
##Initialize model inputs
model_path=r'./model_files/'   #path to the model files

path_to_test_data = r'../Software for qualitative assesment/test_data_real'  #path to the data to be segmented (nrrds)

save_path =r'./patient_segmentations' #path for the output files (nrrds)

pat_dict = {
    0: ['../Software for qualitative assesment/test_data_real/YG_HWSCOBELFJHZ_YG_DGNWZ1SLBAU5_06_26_2019/YG_HWSCOBELFJHZ_YG_DGNWZ1SLBAU5_06_26_2019.nrrd'],
    1: ['../Software for qualitative assesment/test_data_real/YG_HWSCOBELFJHZ_YG_7EBLNXLZDNOY_10_20_2021/YG_HWSCOBELFJHZ_YG_7EBLNXLZDNOY_10_20_2021.nrrd'],
    2: ['../Software for qualitative assesment/test_data_real/YG_HWSCOBELFJHZ_YG_5QRI1MX3J83R_12_19_2022/YG_HWSCOBELFJHZ_YG_5QRI1MX3J83R_12_19_2022.nrrd'],
}

#initialize the model
model = cp(model_path,path_to_test_data,save_path,verbosity=True
           ,pat_dict=pat_dict
          )    #set verbosity =True to see what is going on
```
